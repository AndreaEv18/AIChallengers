```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("mushroom_data_all.csv")
```

```python
X=train.drop('class_edible',axis=1)
y=train['class_edible']
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder

class_mapping={'p': 0, 'e': 1}
train['class_edible'] = train['class_edible'].map(class_mapping)

train.drop("veil-type",axis=1,inplace=True)
train.drop("veil-color",axis=1,inplace=True)

CodificatoreX = LabelEncoder() 
for columns in train.columns:
    train[columns] = CodificatoreX.fit_transform(train[columns])
CodificatoreY=LabelEncoder()
y = CodificatoreY.fit_transform(y)
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split

y = train["class_edible"].values   
x = train.drop(["class_edible"], axis=1).values

x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

decisiontree = DecisionTreeClassifier()
decisiontree.fit(x_train, y_train)
y_pred = decisiontree.predict(x_test)

print("ACCURACY: {}%".format(round(decisiontree.score(x_test,y_test)*100,2)))
```
